```python
# import libraries 

from bs4 import BeautifulSoup as bs
import requests
import time
import datetime
import os

import smtplib
```

```python
#Connect to URL
URL ='https://www.goldsmiths.co.uk/c/Jewellery/Gold-Jewellery'
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36"}
response = requests.get(URL, headers =headers)
src = response.text
```

```python
#save the site page (to avoid blocking)
with open("index.html",'w') as file:
    file.write(src)
```

```python
#read the saved page
with open("index.html") as file:
    src = file.read()
```

```python
#read data
soup = bs(src, "lxml")
```

```python
#extract the data we need
all_brand = soup.find_all(class_ = "productTileBrand")
all_title = soup.find_all(class_ = "productTileName")
all_price = soup.find_all(class_ = "productTilePrice")
```

```python
#Save data into lists
brand, title, price, loan  = [],[],[],[]
for i in all_brand:
    brand.append( i.text)
for i in all_title:
    title.append( i.text)
for i in all_price:
    price.append( i.text)
```

```python
# importing pandas
import pandas as pd

# creating a DataFrame
data = {
   'Brand': brand,
   'Title': title,
   'price': price,
}
dataframe = pd.DataFrame(data)
```

```python
# Clean up the data

#Remove characters
dataframe['price'] = dataframe['price'].str.replace('\n\t\t\t\t\t\t\t\t\t\t', ' ')
dataframe['price'] = dataframe['price'].str.replace('\n', '')
```

```python
# Split column price and remove it
dataframe[['Current Price','Old Price']] = dataframe.price.str.split(" ",expand=True)
dataframe.drop(['price'], axis=1, inplace = True)
```

```python
#Remove characters
dataframe['Current Price'] = dataframe['Current Price'].str.replace('£', '')
dataframe['Old Price'] = dataframe['Old Price'].str.replace('£', '')
```

```python
# Create a Timestamp to track prices
import datetime
today = datetime.date.today()

dataframe['Date'] = today
```

```python
#Check what we got after modifications
dataframe
```

```python
# Create CSV and import dataset there
import csv 

dataframe.to_csv('goldsmiths.csv', index=False)
```

```python
#Now we are appending data to the csv file

with open('goldsmiths.csv', 'a+', newline='') as f:
    dataframe.to_csv(f, header=False,index=False)
```
